```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def volume(x):
    return x * (1 - 2*x)**2

x_values = np.linspace(0, 0.5, 100)

v_values = volume(x_values)

max_v = max(v_values)
max_x = x_values[v_values.argmax()]

user_x = float(input("Digite o valor de x para encontrar o volume correspondente: "))

user_v = volume(user_x)

print(f'O maior valor possível para V é {max_v} quando x é {max_x}')
print(f'Para x = {user_x}, V é {user_v}')

plt.plot(x_values, v_values, label='V(x) = x(1 - 2x)^2')
plt.scatter(max_x, max_v, color='red', label='Máximo')
plt.scatter(user_x, user_v, color='blue', label=f'Para x = {user_x}')

plt.xlabel('Altura (x)')
plt.ylabel('Volume (V)')
plt.title('Gráfico de V(x) em relação à altura')

plt.legend()

plt.show()
```
