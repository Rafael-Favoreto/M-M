```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, integrate
```

```python

x = symbols('x')

def area_retangulo(x):
    return -3/4 * x**2 + 6*x

def area_triangulo(x):
    return 0.5 * 6 * 8

x_user = float(input("Digite o valor de d(D,B) (entre 0 e 8): "))

x_user = max(0, min(8, x_user))

area_retangulo_user = area_retangulo(x_user)
area_triangulo_user = area_triangulo(x_user)

x_values = np.linspace(0, 8, 100)

y_retangulo = [area_retangulo(val) for val in x_values]

plt.plot(x_values, y_retangulo, label='Área do Retângulo')
plt.scatter(x_user, area_retangulo_user, color='red', label=f'Valor inserido: {x_user}')
plt.xlabel('x')
plt.ylabel('Área')
plt.title('Área do Retângulo e do Triângulo em relação a x')
plt.legend()
plt.grid(True)
plt.show()

print(f'Área do Retângulo para x = {x_user}: {area_retangulo_user}')
```
