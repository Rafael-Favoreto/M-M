```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def parabola(x):
    return -x**2 + 60*x

x_values = np.linspace(0, 60, 100)
y_values = parabola(x_values)

vertex_x = 30
vertex_y = parabola(vertex_x)

user_x = float(input("Digite um valor para x: "))
user_y = parabola(user_x)

plt.plot(x_values, y_values, label='y = -x² + 60x')
plt.scatter(vertex_x, vertex_y, color='red', label=f'Vértice ({vertex_x}, {vertex_y})')
plt.scatter(user_x, user_y, color='blue', label=f'Ponto de Usuário ({user_x}, {user_y})')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da Função Quadrática')
plt.legend()

plt.show()
```
